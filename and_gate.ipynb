# AND Gate Using CNN

## Import packages

```python
import numpy as np
from keras.layers import Dense
from keras.models import Sequential
```

## Initializing train and test data

```python
train_data = np.array([[0, 0], [0, 1],
                        [1, 0], [1, 1]])
target_data = np.array([[0], [0], 
                        [0], [1]])
```

## Keras Models
### Adding sequential models

```python
model = Sequential()
model.add(Dense(32, input_dim=2, 
                activation='relu'))
model.add(Dense(1, 
                activation='sigmoid'))
model.summary()
```

### Compiling the model

```python
model.compile(optimizer='adam', loss='mean_squared_error', metrics=['accuracy'])
```

### Training the model

```python
epoch = 500
history = model.fit(train_data, target_data, epochs=epoch, callbacks=None)
```

## Evaluation

```python
scores = model.evaluate(train_data, target_data)
```

## Plotting model

```python
import matplotlib.pyplot as plt
plt.figure(figsize=(4, 4))
plt.plot(range(epoch), history.history['accuracy'], label="Accuracy")
plt.plot(range(epoch), history.history['loss'], label="Loss")
plt.legend(loc="right")
plt.show()
```

## Predict Model

```python
model.predict(np.array([[1, 1]])).round()
```
